Объяснимый искусственный интеллект (XAI)
- Объяснимый искусственный интеллект (XAI) включает в себя набор процедур и техник.
- Его основная цель - облегчить получение понятных и надежных результатов от алгоритмов машинного обучения для обычных пользователей.
- XAI служит ключевым элементом в парадигме справедливости, ответственности и прозрачности (FAT) машинного обучения.
- Он решает потребность в прозрачности и интерпретируемости в системах искусственного интеллекта.
- XAI часто обсуждается в контексте глубокого обучения, которое часто включает в себя сложные и замысловатые архитектуры нейронных сетей.
- Его применение в глубоком обучении направлено на то, чтобы сделать эти сложные модели более интерпретируемыми.
- Организации, внедряющие искусственный интеллект, стремятся установить доверие к используемой технологии.
- XAI играет важную роль в этом процессе, позволяя пользователям понимать поведение моделей искусственного интеллекта.

Техники объяснимого искусственного интеллекта (XAI)

**LIME (Local Interpretable Model-agnostic Explanations):**
- **Подход:** LIME - широко используемая техника XAI, которая создает локальную модель для объяснения интерпретируемости глобальной модели.
- **Реализация на Python:** Используйте пакет 'lime', который предоставляет инструменты и функции для генерации и интерпретации объяснений LIME.
- **Цель:** LIME предоставляет понимание факторов, наиболее значимых в локальном контексте конкретного прогноза.

**SHAP (SHapley Additive exPlanations):**
- **Подход:** SHAP использует значения Шепли из теории игр для предоставления интерпретируемых идей о факторах, влияющих на прогнозы модели.
- **Реализация на Python:** Пакет 'shap' предоставляет обширный набор инструментов и функций для реализации и интерпретации объяснений SHAP.
- **Цель:** Объяснения SHAP раскрывают вклад каждого признака в общий прогноз модели.

**ELI5 (Explain Like I’m 5):**
- **Подход:** ELI5 упрощает сложные объяснения модели, используя интуитивный язык, что делает их доступными для неспециалистов.
- **Реализация на Python:** Пакет 'eli5' предлагает инструменты и функции для генерации и интерпретации объяснений ELI5 удобным для пользователя способом.
- **Цель:** ELI5 предоставляет интерпретируемые идеи о факторах, влияющих на прогнозы модели, простым и понятным образом.

## LIME (Local Interpretable Model-agnostic Explanations)

LIME - это метод интерпретации моделей машинного обучения, который стремится сделать прогнозы моделей более понятными на уровне конкретных предсказаний. Он является модельно-независимым, что означает, что его можно применять к любой модели, независимо от её типа.

### Принцип работы LIME:

1. **Выбор экземпляра:**
   LIME начинает с выбора конкретного экземпляра данных, для которого требуется объяснение. Этот экземпляр может быть как положительным, так и отрицательным по отношению к модели.

2. **Генерация аналогов:**
   Затем создаются "аналоги" выбранного экземпляра, которые являются схожими с ним по признакам. Эти аналоги используются для обучения локальной модели.

3. **Обучение локальной модели:**
   LIME обучает локальную модель (часто линейную), используя аналоги. Эта модель приближает поведение исходной модели в окрестности выбранного экземпляра.

4. **Вычисление вкладов признаков:**
   После обучения локальной модели LIME вычисляет вклад каждого признака в предсказание для выбранного экземпляра. Это позволяет интерпретировать, какие признаки оказывают наибольшее влияние на конкретное предсказание.

### Преимущества LIME:

- **Модельная независимость:** Может применяться к любым моделям машинного обучения.
- **Локальность:** Предоставляет интерпретацию для конкретных предсказаний, а не всей модели.

### Ограничения LIME:

- **Определение аналогов:** Качество интерпретации зависит от того, насколько успешно подобраны аналоги.
- **Ограничения линейной модели:** Интерпретация основана на линейной модели, что может быть ограничивающим в случае нелинейных моделей.

````python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
````

````python
columns = [
    'Class', 'Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash',
    'Magnesium', 'Total phenols', 'Flavanoids', 'Nonflavanoid phenols',
    'Proanthocyanins', 'Color intensity', 'Hue', 'OD280/OD315 of diluted wines', 'Proline'
]
df = pd.read_csv('Datasets/wine.csv', header=None, names=columns)
df.head()
````

````python
sns.set(style="whitegrid")
variables_to_plot = ['Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash', 'Magnesium', 'Total phenols', 'Flavanoids',
                     'Nonflavanoid phenols', 'Proanthocyanins', 'Color intensity', 'Hue',
                     'OD280/OD315 of diluted wines']
````

````python
df.columns = df.columns.str.strip()
````

````python
num_variables = len(variables_to_plot)
num_rows = (num_variables - 1) // 3 + 1
num_cols = min(num_variables, 3)
````

````python
fig, axes = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(15, 15))
axes = axes.flatten()

for i, variable in enumerate(variables_to_plot):
    sns.kdeplot(df[variable], ax=axes[i], color='salmon', fill=True, common_norm=False)
    axes[i].set_title(f'Distribution of {variable}')
    axes[i].set_xlabel(f'Values of {variable}')
    axes[i].set_ylabel('Density')
plt.tight_layout(h_pad=0.8)
plt.show()
````

````python
x = df.drop(columns=['Class'])
y = df['Class']
````

````python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, random_state=42)
````

````python
clf = RandomForestClassifier(n_estimators=100, max_depth=5, min_samples_split=2, min_samples_leaf=1, random_state=42)
clf.fit(x_train, y_train)
````

````python
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 5, 7],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
grid_search = GridSearchCV(clf, param_grid, cv=5, verbose=10)
grid_search.fit(x_train, y_train)
````

````python
clf = grid_search.best_estimator_
````

````python
y_pred = clf.predict(x_test)
report = classification_report(y_test, y_pred)
print(report)
````

````python
import lime
from lime import lime_tabular

lime_explainer = lime_tabular.LimeTabularExplainer(training_data=np.array(x_train), feature_names=x_train.columns, class_names=['1', '2', '3'], mode='classification')
````

````python
explanation = lime_explainer.explain_instance(data_row=x_test.iloc[1], predict_fn=clf.predict_proba, top_labels=6, num_features=13)
explanation.show_in_notebook()
````
