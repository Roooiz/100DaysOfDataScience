Объяснимый искусственный интеллект (XAI)
- Объяснимый искусственный интеллект (XAI) включает в себя набор процедур и техник.
- Его основная цель - облегчить получение понятных и надежных результатов от алгоритмов машинного обучения для обычных пользователей.
- XAI служит ключевым элементом в парадигме справедливости, ответственности и прозрачности (FAT) машинного обучения.
- Он решает потребность в прозрачности и интерпретируемости в системах искусственного интеллекта.
- XAI часто обсуждается в контексте глубокого обучения, которое часто включает в себя сложные и замысловатые архитектуры нейронных сетей.
- Его применение в глубоком обучении направлено на то, чтобы сделать эти сложные модели более интерпретируемыми.
- Организации, внедряющие искусственный интеллект, стремятся установить доверие к используемой технологии.
- XAI играет важную роль в этом процессе, позволяя пользователям понимать поведение моделей искусственного интеллекта.

Техники объяснимого искусственного интеллекта (XAI)

**LIME (Local Interpretable Model-agnostic Explanations):**
- **Подход:** LIME - широко используемая техника XAI, которая создает локальную модель для объяснения интерпретируемости глобальной модели.
- **Реализация на Python:** Используйте пакет 'lime', который предоставляет инструменты и функции для генерации и интерпретации объяснений LIME.
- **Цель:** LIME предоставляет понимание факторов, наиболее значимых в локальном контексте конкретного прогноза.

**SHAP (SHapley Additive exPlanations):**
- **Подход:** SHAP использует значения Шепли из теории игр для предоставления интерпретируемых идей о факторах, влияющих на прогнозы модели.
- **Реализация на Python:** Пакет 'shap' предоставляет обширный набор инструментов и функций для реализации и интерпретации объяснений SHAP.
- **Цель:** Объяснения SHAP раскрывают вклад каждого признака в общий прогноз модели.

**ELI5 (Explain Like I’m 5):**
- **Подход:** ELI5 упрощает сложные объяснения модели, используя интуитивный язык, что делает их доступными для неспециалистов.
- **Реализация на Python:** Пакет 'eli5' предлагает инструменты и функции для генерации и интерпретации объяснений ELI5 удобным для пользователя способом.
- **Цель:** ELI5 предоставляет интерпретируемые идеи о факторах, влияющих на прогнозы модели, простым и понятным образом.

Одним из инструментов в области XAI является SHAP (SHapley Additive exPlanations) – метод объяснения моделей машинного обучения, основанный на теории игр. Этот метод предоставляет оценки вклада каждого признака в конечное предсказание модели. Давай подробнее разберем, как работает SHAP.

**SHAP (SHapley Additive exPlanations):**

1. **Идея:**
   SHAP базируется на концепции "значения Шепли" из теории игр. Значения Шепли представляют собой способ распределения выигрыша между участниками коалиции в кооперативных играх. В контексте машинного обучения, участники – это признаки, а "выигрыш" – это предсказание модели.

2. **Оценка вклада признаков:**
   SHAP оценивает вклад каждого признака в предсказание, учитывая все возможные комбинации признаков. Это делается путем вычисления средневзвешенных вкладов каждого признака по всем возможным перестановкам признаков.

3. **Положительные и отрицательные вклады:**
   Положительный вклад признака указывает, насколько его присутствие увеличивает предсказание, в то время как отрицательный вклад указывает на уменьшение предсказания. SHAP раскрывает, как каждый признак вносит свой вклад в итоговый результат.

4. **Важность признаков:**
   На основе оценок вклада SHAP можно определить важность каждого признака в предсказаниях модели. Это позволяет аналитикам и разработчикам лучше понять, какие факторы оказывают наибольшее влияние на модель.

5. **Библиотека SHAP:**
   Существует библиотека SHAP для различных платформ и языков программирования, включая Python. Она предоставляет удобные функции для вычисления SHAP значений и визуализации результатов.

````python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import shap

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
````

````python
columns = [
    'Class', 'Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash',
    'Magnesium', 'Total phenols', 'Flavanoids', 'Nonflavanoid phenols',
    'Proanthocyanins', 'Color intensity', 'Hue', 'OD280/OD315 of diluted wines', 'Proline'
]
df = pd.read_csv('Datasets/wine.csv', header=None, names=columns)
df.head()
````

````python
sns.set(style="whitegrid")

# Список переменных для построения графика
variables_to_plot = ['Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash', 'Magnesium',
                     'Total phenols', 'Flavanoids', 'Nonflavanoid phenols', 'Proanthocyanins',
                     'Color intensity', 'Hue', 'OD280/OD315 of diluted wines']

# Удалите лишний пробелы в названиях столбцов
df.columns = df.columns.str.strip()

# Вычислим количество строк и столбцов для сетки
num_variables = len(variables_to_plot)
num_rows = (num_variables - 1) // 3 + 1
num_cols = min(num_variables, 3)

# Создадим подзаголовки с динамической сеткой и вертикальным интервалом
fig, axes = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(15, 15))

# Выровняем оси для удобства итерации
axes = axes.flatten()

# Перебирём переменные и выведем на каждый подраздел
for i, variable in enumerate(variables_to_plot):
    sns.kdeplot(df[variable], ax=axes[i], color='blue', fill=True, common_norm=False)
    axes[i].set_title(f'Distribution of {variable}')
    axes[i].set_xlabel(f'Values of {variable}')
    axes[i].set_ylabel('Density')

plt.tight_layout(h_pad=1)
plt.show()
````

# Создадим модель

````python
# Разделим данные на независимые и зависимые переменные
x = df.drop(columns=['Class'])
y = df['Class']
````

````python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, random_state=42)
````

````python
clf = RandomForestClassifier()
clf.fit(x_train, y_train)
````

````python
# создадим словарь и список значений для оптимизации модели
params = {
    'n_estimators': [100, 500, 1000],
    'criterion': ['gini', 'entropy'],
    'max_depth': [3, 4, 5, 6, 7, 8, 9, 10],
}
````

````python
# Сделаем поиск по сетке для поиска наиболее оптимальных параметров
grid_search = GridSearchCV(clf, params, cv=10, verbose=10)
````

````python
grid_search.fit(x_train, y_train)
````

````python
# Выберем лучшую модель
clf = grid_search.best_estimator_
````

````python
# Сделаем прогноз по тестовому набору данных
y_pred = clf.predict(x_test)
````

````python
report = classification_report(y_test, y_pred)
print(report)
````

# Объяснимость с помощью SHAP

````python
# Создадим экземпляр класса TreeExplainer
explainer = shap.TreeExplainer(clf)
````

````python
# Получим корректные значения данных
shap_values = explainer.shap_values(x)
````

````python
# Построим график зависимости значений shap от алкоголя
shap.dependence_plot('Alcohol', shap_values[0], x)
````

````python
# Построим график зависимости значений shap от всех объектов
shap.summary_plot(shap_values[0], x)
````

````python
shap.summary_plot(shap_values[1], x)
````
