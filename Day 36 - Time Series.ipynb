### Временные ряды

#### Определение:
Временные ряды представляют собой последовательность данных, измеренных или наблюдаемых в разные моменты времени. Это могут быть замеры температуры каждый час, ежедневные финансовые показатели, количество проданных товаров каждый месяц и т. д. Основная особенность временных рядов - это их упорядоченность во времени.

#### Примеры:
- **Ежедневные цены на акции:** График, отображающий изменения стоимости акции компании каждый день.
  
- **Ежемесячные продажи в магазине:** Статистика, отражающая количество проданных товаров в магазине каждый месяц.

### 2. Характеристики временных рядов:

#### Временная компонента:
Временные ряды отражают динамику данных во времени. Например, если мы рассматриваем годовой временной ряд температуры, мы можем наблюдать сезонные изменения, такие как повышение температуры летом и снижение зимой.

#### Тренд:
Тренд представляет собой общую направленность изменений в долгосрочной перспективе. Например, если у временного ряда по продажам товаров есть положительный тренд, это может указывать на рост популярности продукции.

#### Сезонность:
Сезонность отражает повторяющиеся циклические изменения в данных. Вернемся к примеру с температурой: сезонность проявляется в изменениях температуры в зависимости от времени года, что создает ярко выраженные сезоны.

### 3. Стационарность временных рядов:

#### Определение стационарности:
Стационарность временного ряда важна для анализа, поскольку она предполагает постоянство статистических свойств данных во времени. Стационарные временные ряды обладают постоянством среднего и дисперсии, что упрощает их анализ. Если временной ряд стационарен, его прошлые статистические свойства могут быть использованы для прогнозирования будущих.

#### Тестирование стационарности:
Простым способом проверки стационарности является визуальный анализ графика временного ряда и использование статистических тестов. Если график показывает явный тренд или сезонные изменения, это может свидетельствовать о нестационарности. Статистические тесты, такие как тест Дики-Фуллера, помогают формально оценить стационарность.

### 4. Модели временных рядов:

#### ARIMA-модели:
ARIMA (Autoregressive Integrated Moving Average) - это класс моделей временных рядов, объединяющий авторегрессию (AR), интегрирование (I) и скользящее среднее (MA). 

- **Авторегрессия (AR):** Элементы ряда зависят от предыдущих значений. Например, текущая температура может зависеть от предыдущих дней.

- **Интегрирование (I):** Процесс вычитания текущего значения ряда из предыдущего для обеспечения стационарности.

- **Скользящее среднее (MA):** Элементы ряда зависят от случайных шоковых значений.

ARIMA-модели позволяют учитывать тренды и сезонность, делая прогнозы более точными.

#### Сезонные модели:
Для учета сезонности используются модели, которые улавливают повторяющиеся циклические изменения в данных. Это могут быть SARIMA (Seasonal ARIMA) или другие модели, специфичные для сезонных данных.

#### Экспоненциальное сглаживание:
Экспоненциальное сглаживание - это метод, основанный на взвешенном учете предыдущих значений временного ряда. Существуют различные методы экспоненциального сглаживания, такие как простое экспоненциальное сглаживание (SES) и двойное экспоненциальное сглаживание (Holt-Winters). SES учитывает только последние значения, в то время как Holt-Winters учитывает тренд и сезонность.

### 5. Прогнозирование и анализ временных рядов:

#### Тренировка и тестирование моделей:
Для построения надежной модели временного ряда, данные обычно разделяют на две части: обучающую и тестовую выборки. Модель обучается на обучающих данных, после чего ее эффективность проверяется на тестовых данных. Это позволяет оценить, насколько хорошо модель обобщает знания на новых данных.

#### Оценка точности прогнозов:
Для измерения точности прогнозов используются различные метрики, такие как среднеквадратичная ошибка (MSE) или средняя абсолютная ошибка (MAE). Эти метрики позволяют оценить расхождение между фактическими и прогнозируемыми значениями и определить эффективность модели.

### 6. Инструменты для анализа временных рядов:
Для анализа временных рядов часто применяются в Python - Statsmodels, а в R - forecast и timeSeries. Эти инструменты предоставляют различные методы анализа и моделирования временных рядов.

### Формулы, которые могут пригодиться в анализе временных рядов:

#### 1. **Простое экспоненциальное сглаживание (SES):**
   - Прогноз на следующий период: 
     $ \hat{Y}_{t+1} = \alpha \cdot Y_t + (1-\alpha) \cdot \hat{Y}_t $
   где $ Y_t $ - текущее значение, $ \hat{Y}_t $ - прогноз, $ \alpha $ - коэффициент сглаживания (0 < $ \alpha $ < 1).

#### 2. **ARIMA (p, d, q):**
   - Уравнение ARIMA:
     $ Y_t = c + \phi_1 \cdot Y_{t-1} + \phi_2 \cdot Y_{t-2} + \ldots + \phi_p \cdot Y_{t-p} + \varepsilon_t - \theta_1 \cdot \varepsilon_{t-1} - \theta_2 \cdot \varepsilon_{t-2} - \ldots - \theta_q \cdot \varepsilon_{t-q} $
   где $ p $ - порядок авторегрессии, $ d $ - порядок интегрирования, $ q $ - порядок скользящего среднего.

#### 3. **Сезонная компонента в SARIMA (p, d, q)(P, D, Q)$ _s $:**
   - Уравнение SARIMA:
     $ Y_t = c + \Phi_1 \cdot Y_{t-s} + \phi_1 \cdot Y_{t-1} + \phi_2 \cdot Y_{t-2} + \ldots + \phi_p \cdot Y_{t-p} + \varepsilon_t - \Theta_1 \cdot \varepsilon_{t-s} - \theta_1 \cdot \varepsilon_{t-1} - \theta_2 \cdot \varepsilon_{t-2} - \ldots - \theta_q \cdot \varepsilon_{t-q} $
   где $ P $, $ D $, $ Q $ - порядки сезонной составляющей, а $ s $ - период сезонности.

#### 4. **Средняя абсолютная ошибка (MAE):**
   $ MAE = \frac{1}{n} \sum_{i=1}^{n} |Y_i - \hat{Y}_i| $

#### 5. **Среднеквадратичная ошибка (MSE):**
   $ MSE = \frac{1}{n} \sum_{i=1}^{n} (Y_i - \hat{Y}_i)^2 $

#### 6. **Тест Дики-Фуллера для проверки стационарности:**
   Гипотеза теста:
   - $ H_0 $ : Ряд нестационарен.
   - $ H_1 $ : Ряд стационарен.

   Если p-value меньше уровня значимости, отклоняем нулевую гипотезу о нестационарности.

# Анализ временных рядов

## Проверим на стационарность

```python
from statsmodels.tsa.stattools import  adfuller


timeseries = df['Passengers']


def stationarity_test(timeseries):
    rolling_mean = timeseries.rolling(window=12).mean()
    rolling_std = timeseries.rolling(window=12).std()

    plt.figure(figsize=(10, 6))
    plt.xlabel('Годы')
    plt.ylabel('Кол-во пассажиров')
    plt.title('Стационарный тест: Скользящее среднее и стандартное отклонение')
    plt.plot(timeseries, color='blue', label='Оригинал')
    plt.plot(rolling_mean, color='green', label='Скользящее среднее')
    plt.plot(rolling_std, color='red', label='Скользящее стандартное отклонение')
    plt.legend()
    plt.show()

    # Тест Дики-Фуллера
    print('Результаты теста Дики-Фуллера')
    df_test = adfuller(timeseries)
    df_output = pd.Series(df_test[0:4],
                          index=['Тестовая статистика', 'p-value', 'Лаг', 'Кол-во наблюдений'])
    for key, value in df_test[4].items():
        df_output['Критическое значение (%s)' % key] = value
    print(df_output)

    return rolling_mean, rolling_std


rolling_mean, rolling_std = stationarity_test(timeseries)
```
