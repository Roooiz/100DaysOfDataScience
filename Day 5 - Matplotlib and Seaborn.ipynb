# Стили линий

# Свойства шрифта для заголовка и надписей

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4, 5, 6, 7])
y = np.array([240, 250, 260, 270, 280, 350, 130])

font1 = {'family': 'monospace', 'color': 'blue', 'size': 20}
font2 = {'family': 'serif', 'color': 'darkred', 'size': 15}
font3 = {'family': 'sans-serif', 'color': 'green', 'size': 15}

plt.title("Акции BimBim", fontdict=font1)
plt.xlabel("Дата", fontdict=font2)
plt.ylabel("Стоимость", fontdict=font3)

plt.plot(x, y)
plt.show()
```

# Расположение заголовка
Мы можем использовать параметр loc в title() для размещения заголовка.
Допустимыми значениями являются: ‘left’, ‘right’ и ‘center’. Значение по умолчанию - "center’.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4, 5, 6, 7])
y = np.array([240, 250, 260, 270, 280, 350, 130])

plt.plot(x, y)

plt.title("Акции BimBim", loc="right")
plt.xlabel("Дата")
plt.ylabel("Стоимость")

plt.show()
```

# Размещение нескольких графиков subplot()

Функция subplot() принимает три аргумента, описывающих расположение рисунка.
Расположение организовано в строки и столбцы, которые представлены первым и вторым аргументом. Третий аргумент представляет индекс текущего графика.
**plt.subplot(1, 2, 1)**    # график на 1 строке, 2 столбца, и этот график является первым.
**plt.subplot(1, 2, 2)**    # график на 1 строке, 2 столбца, и этот график является вторым графиком.

```python
import matplotlib.pyplot as plt
import numpy as np

# График 1
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x, y)

# График 2
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x, y)

plt.show()
```

# Точечный график Scatterplot

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4, 5])
y = np.array([11, 16, 4, 2, 7])
plt.scatter(x, y)

plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.title('Точечный график')
plt.show()
```

# Покрасим наши точки

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4, 5])
y = np.array([11, 16, 4, 2, 7])
colors = np.array(["red", "green", "blue", "pink", "gray"])
plt.scatter(x, y, c=colors)
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.title('Точечный график (раскрашенные точки)')
plt.show()
```

# Добавим colormap

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4, 5])
y = np.array([11, 16, 4, 2, 7])
colors = np.array([0, 10, 20, 30, 40])
plt.scatter(x, y, c=colors, cmap='viridis')
plt.colorbar()
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.title('Точечный график c colormap')
plt.show()
```

# Размерности точек

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4, 5])
y = np.array([11, 16, 4, 2, 7])
sizes = np.array([20, 50, 100, 200, 500])
plt.scatter(x, y, alpha=0.8, s=sizes)

plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.title('Точечный график')
plt.show()
```

# Столбчатая диаграмма Barplot

```python
# Создание диаграммы

import matplotlib.pyplot as plt
import numpy as np

x = np.array(['BimBim', 'BamBam', 'LitIT', 'MosTech'])
y = np.array([25, 40, 30, 50])

plt.bar(x, y)

plt.xlabel('Компании')
plt.ylabel('Выручка, млрд.руб')
plt.title('Доходы')
plt.show()
```

```python
# Горизонтальный график

import matplotlib.pyplot as plt
import numpy as np

x = np.array(['BimBim', 'BamBam', 'LitIT', 'MosTech'])
y = np.array([25, 40, 30, 50])

plt.barh(x, y, color="yellow")  # Для горизонтального графика используем plt.barh()

plt.xlabel('Кол-во сотрудников')
plt.ylabel('Компании')
plt.title('Сотрудники')
plt.show()
```

# Ширина столбцов
Функция bar() принимает аргумент width, чтобы задать ширину полос на графике.
Для горизонтальных полос используем высоту вместо ширины (ЛОГИЧНО).

```python
# Создание диаграммы

import matplotlib.pyplot as plt
import numpy as np

x = np.array(['BimBim', 'BamBam', 'LitIT', 'MosTech'])
y = np.array([25, 40, 30, 50])

plt.bar(x, y, width=0.2)

plt.xlabel('Компании')
plt.ylabel('Выручка, млрд.руб')
plt.title('Доходы')
plt.show()
```

# Гистограммы Histogram

Гистограмма - это график, показывающий распределение частот.
Это график, показывающий количество наблюдений в пределах каждого заданного интервала.
Пример: Допустим, вы решили рассмотреть распределение роста 300 человек.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170, 10, 300)

plt.hist(x)
plt.show()
```

# Круговые диаграммы Pie Chart

```python
# Создание круговой диаграммы
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
companies = ["Apple", "Microsoft", "Amazon", "Netflix"]

plt.pie(y, labels=companies, startangle=90)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
companies = ["Apple", "Microsoft", "Amazon", "Netflix"]
explod = [0.2, 0, 0, 0]  # Разрыв круговой диаграммы

plt.pie(y, labels=companies, explode=explod, shadow=True)  # shadow = тень
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
companies = ["Apple", "Microsoft", "Amazon", "Netflix"]
explod = [0.1, 0.3, 0.1, 0.1]  # Разрыв круговой диаграммы

plt.pie(y, labels=companies, explode=explod, shadow=True)  # shadow = тень
plt.show()
```

```python
# Добавим легенду
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
companies = ["Apple", "Microsoft", "Amazon", "Netflix"]
color = ["black", "hotpink", "b", "#4CAF50"]

plt.pie(y, labels=companies, colors=color)
plt.legend(title="Топ компании")
plt.show() 
```
