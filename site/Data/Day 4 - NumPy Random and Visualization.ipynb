# NumPy Random

Случайное число - число, которое невозможно предсказать логически.

```python
from numpy import random

x1 = random.rand()  # метод rand() возвращает случайное значение с плавающей точкой от 0 до 1.
x2 = random.rand(5)  # Сгенерируем одномерный массив, содержащий 5 случайных значений с плавающей точкой.
x3 = random.rand(3, 5)  # Сгенерируем двумерный массив из 3 строк, каждая строка содержит 5 случайных чисел.
x4 = random.choice([3, 5, 7, 9])  # Метод choice() позволяет генерировать случайное значение на основе массива значений.
x5 = random.choice([3, 5, 7, 9],
                   size=(3, 5))  # Метод choice() также позволяет возвращать массив значений.Добавим параметр размера.
x6 = random.randint(100)  # Сгенерируем случайное целое число от 0 до 100.
x7 = random.randint(100,
                    size=(5))  # Метод randint() принимает параметр размера, в котором вы можете указать форму массива.
x8 = random.randint(100, size=(
    3, 5))  # Сгенерируем двумерный массив из 3 строк, каждая строка содержит 5 случайных целых чисел от 0 до 100.

arr_x = [x1, x2, x3, x4, x5, x6, x7, x8]
for x in arr_x:
    print(x)
    print("-" * 25)
```

# Распределения
Первое и самое важное распределение - нормальное (Распределение Гаусса). Оно соответствует распределению вероятностей многих событий, например. Показатели IQ, сердцебиение и т.д.
Также существует множетство других распределений:
1. Биномиальное распределение
2. Распределение Пуассона
3. Равномерное распределение
4. Логистическое распределение
5. Мультиномиальное распределение
6. Экспоненциальное распределение
7. Распределение Хи-квадрат
8. Распределение Рэлея
9. Распределение Парето
10. Распределение Ципфа

Рассмотрим каждое из них

```python
from numpy import random

'''
НОРМАЛЬНОЕ РАСПРЕДЕЛЕНИЕ

У него есть три параметра:
1. loc - (среднее значение), где находится пик колокола.
2. scale - (стандартное отклонение), насколько плоским должно быть распределение графика.
3. size - форма возвращаемого массива.
'''
y0 = random.normal(loc=1, scale=2, size=(2, 3))
#-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

'''
БИНОМИАЛЬНОЕ РАСПРЕДЕЛЕНИЕ
Биномиальное распределение - это дискретное распределение.
Оно описывает результат бинарных сценариев, например, при подбрасывании монеты выпадет либо орел, либо решка.
Оно имеет три параметра:
1. n - количество испытаний.
2. p - вероятность выполнения каждого испытания (например, для подбрасывания монеты по 0,5).
3. size - форма возвращаемого массива.
'''
y1 = random.binomial(n=10, p=0.5, size=10)
#-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

'''
РАСПРЕДЕЛЕНИЕ ПУАССОНА
Распределение Пуассона является дискретным распределением.
Оно оценивает, сколько раз событие может произойти за заданное время. Например, если кто-то ест два раза в день, какова вероятность того, что он будет есть трижды?
Оно имеет два параметра:
1. lam - частота или известное количество вхождений, например, 2 для вышеуказанной задачи.
2. size - форма возвращаемого массива.
'''
y2 = random.poisson(lam=2, size=10)
#-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

'''
РАВНОМЕРНОЕ РАСПРЕДЕЛЕНИЕ
Используется для описания вероятности, при которой каждое событие имеет равные шансы произойти. Например, генерация случайных чисел.
Имеет три параметра:
1. a - нижняя граница - значение по умолчанию 0.0.
2. b - верхняя граница - значение по умолчанию 1.0.
3. size - форма возвращаемого массива.
'''
y3 = random.uniform(size=(2, 3))
#-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

'''
ЛОГИСТИЧЕСКОЕ РАСПРЕДЕЛЕНИЕ
Логистическое распределение используется для описания роста. Широко используется в машинном обучении при логистической регрессии, нейронных сетях и т.д.
Оно имеет три параметра:
1. loc - среднее значение, где находится пик. По умолчанию 0.
2. scale - стандартное отклонение, равномерность распределения. По умолчанию 1.
3. size - форма возвращаемого массива.
'''
y4 = random.logistic(loc=1, scale=2, size=(2, 3))
#-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

'''
МУЛЬТИНОМИАЛЬНОЕ РАСПРЕДЕЛЕНИЕ
Мультиномиальное распределение является обобщением биномиального распределения.
Оно описывает исходы мультиномиальных сценариев в отличие от биномиальных, где сценарии должны быть только одним из двух. Например, группа крови населения, результат броска костей.
Он имеет три параметра:
1. n - количество возможных исходов (например, 6 для броска костей).
2. pvals - список вероятностей исходов (например, [1/6, 1/6, 1/6, 1/6, 1/6, 1/6] для броска костей).
3. size - форма возвращаемого массива.
'''
y5 = random.multinomial(n=6, pvals=[1 / 6, 1 / 6, 1 / 6, 1 / 6, 1 / 6, 1 / 6])
#-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

'''
ЭКСПОНЕНЦИАЛЬНОЕ РАСПРЕДЕЛЕНИЕ
Экспоненциальное распределение используется для описания времени до следующего события, например, сбоя/успеха и т.д.
Оно имеет два параметра:
1. scale - обратная величина скорости (см. распределение Ламина Пуассона) по умолчанию равна 1.0.
2. size - форма возвращаемого массива
'''
y6 = random.exponential(scale=2, size=(2, 3))
#-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

'''
РАСПРЕДЕЛЕНИЕ ХИ-КВАДРАТ
Распределение хи-квадрат используется в качестве основы для проверки гипотезы.
Оно имеет два параметра:
1. df - (степень свободы).
2. size - форма возвращаемого массива.
'''
y7 = random.chisquare(df=2, size=(2, 3))
#-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

'''
РАСПРЕДЕЛЕНИЕ РЭЛЕЯ
Распределение Рэлея используется при обработке сигналов.
У него есть два параметра:
1. scale - (стандартное отклонение) определяет, насколько плоским будет распределение (по умолчанию 1.0).
2. size - форма возвращаемого массива.
'''
y8 = random.rayleigh(scale=2, size=(2, 3))
#-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

'''
РАСПРЕДЕЛЕНИЕ ПАРЕТО
Распределение, соответствующее закону Парето, т.е. распределение 80 на 20 (20% факторов приводят к 80% результата).
У него есть два параметра:
1. a - %.
2. size - форма возвращаемого массива
'''
y9 = random.pareto(a=2, size=(2, 3))
#-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

'''
РАСПРЕДЕЛЕНИЕ ЦИПФА
Распределения Ципфа используются для выборки данных на основе закона Ципфа.
У него есть два параметра:
1. a - параметр распределения.
2. size - форма возвращаемого массива
'''
y10 = random.zipf(a=2, size=(2, 3))
#-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------


arr_x = [y0, y1, y2, y3, y4, y5, y6, y7, y8, y9, y10]
for x in arr_x:
    print(x)
    print("-" * 25)
```

# Визуализация распределений

Заранее установим необходимые библиотеки (обе для построения графиков)
**pip install matplotlib**
**pip install seaborn**

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')  # Игнорим все предупреждения и INFO сообщения, чтобы не засорять наш ноутбук

# Создадим фигуру с заданными размерами, куда поместим наши графики
plt.figure(figsize=(12, 12))

# Строим нормальное распределение (Гауссовское)
plt.subplot(331)
sns.distplot(random.normal(size=1000), hist=False)
plt.title('Нормальное распределение')

# Строим биномиальное распределение
plt.subplot(332)
sns.distplot(random.binomial(n=10, p=0.5, size=1000), hist=True, kde=False)
plt.title('Биномиальное распределение')

# Строим распределение Пуассона
plt.subplot(333)
sns.distplot(random.poisson(lam=2, size=1000), kde=False)
plt.title('Распределение Пуассона')

# Строим равномерное распределение
plt.subplot(334)
sns.distplot(random.uniform(size=1000), hist=False)
plt.title('Равномерное распределение')

# Строим логистическое распределение
plt.subplot(335)
sns.distplot(random.logistic(size=1000), hist=False)
plt.title('Логистическое распределение')

# Строим мультиномиальное распределение
plt.subplot(336)
sns.distplot(random.multinomial(n=6, pvals=[1 / 6, 1 / 6, 1 / 6, 1 / 6, 1 / 6, 1 / 6]))
plt.title('Мультиномиальное распределение')

# Строим экспоненциальное распределение
plt.subplot(337)
sns.distplot(random.exponential(size=1000), hist=False)
plt.title('Экспоненциальное распределение')

# Строим распределение Хи-квадрат
plt.subplot(338)
sns.distplot(random.chisquare(df=1, size=1000), hist=False)
plt.title('Распределение Хи-квадрат')

# Строим распределение Рэлея
plt.subplot(339)
sns.distplot(random.rayleigh(size=1000), hist=False)
plt.title('Распределение Рэлея')

plt.tight_layout()
plt.show()  # Отображаем наш итоговый график
```

# Маркеры Matplotlib

Отметим каждую точку кружком

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([3, 8, 1, 10])
plt.plot(y, marker='o')
plt.show()
```

Отметим каждую точку звездочкой

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([3, 8, 1, 10])
plt.plot(y, marker='*')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([3, 8, 1, 10])
plt.plot(y, 'o:b')  # o = точки в форме круга, b = синий цвет (blue), : = точечная линия
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([3, 8, 1, 10])
plt.plot(y, marker='h', color='g')  # h = точки в форме шестиугольника, g = зеленый цвет (green)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([3, 8, 1, 10])
plt.plot(y, marker='p', color='r')  # p = точки в форме пятиугольника, r = красный цвет (red)
plt.show()
```

# Размер маркера и его цвет

Для раскрашивания маркеров matplotlib доступны два параметра:
1. Цвет края маркера Matplotlib (mec).
2. Цвет грани маркера Matplotlib (mfc).

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([3, 8, 1, 10])
plt.plot(y, marker='p', ms=25)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([3, 8, 1, 10])
plt.plot(y, 'X:r', ms=20, mec='black')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([3,8,1,10])
plt.plot(y, 'X:r', ms = 20,mec='black',mfc='blue' )
plt.show()
```
