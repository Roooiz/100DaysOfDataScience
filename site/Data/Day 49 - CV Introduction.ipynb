# Компьютерное зрение

Компьютерное зрение (Computer Vision) представляет собой область исследований в области искусственного интеллекта, направленную на обучение машин анализу и интерпретации визуальной информации. Эта область стремится дать компьютерам способность "видеть" и понимать изображения и видео, как это делает человек.

## Основы компьютерного зрения

### 1. Изображения

Изображение — это представление визуальной информации в форме матрицы пикселей. Пиксель (от "picture element") является самым маленьким элементом изображения и содержит информацию о цвете или яркости.

### 2. Цветовая модель

Изображения могут быть представлены в различных цветовых моделях, таких как RGB (красный, зеленый, синий), HSV (оттенок, насыщенность, яркость) и др. Каждый цвет можно представить как комбинацию базовых цветов в соответствующей модели.

### 3. Формат изображения

Изображения могут быть сохранены в различных форматах, таких как JPEG, PNG, GIF и другие. Каждый формат имеет свои особенности, влияющие на размер файла и качество изображения.

### 4. Обработка изображений

Обработка изображений включает в себя различные техники, такие как фильтрация, улучшение контрастности, обрезка и изменение размера. Они используются для улучшения качества изображений перед их анализом.

### 5. Операции над изображениями

Операции над изображениями включают базовые математические операции над пикселями, такие как сложение, вычитание, умножение и др. Эти операции могут применяться для изменения изображений.

## Обнаружение объектов

### 1. Определение объектов

Обнаружение объектов - это процесс выделения и классификации объектов на изображении. Методы обнаружения могут быть основаны на геометрических признаках, структурных шаблонах или использовать глубокие нейронные сети.

### 2. Сегментация изображений

Сегментация изображений разделяет изображение на более мелкие части, что упрощает анализ каждого объекта отдельно. Это может быть бинарная сегментация (объект/фон) или сегментация по классам.

## Глубокое обучение и Нейронные Сети

### 1. Глубокое обучение

Глубокое обучение - это подраздел машинного обучения, в котором используются глубокие нейронные сети с множеством слоев для обучения на больших объемах данных.

### 2. Нейронные сети

Нейронные сети представляют собой модель, вдохновленную структурой человеческого мозга. Она состоит из нейронов, слоев и весов, и обучается на данных для выполнения задач, таких как классификация, регрессия или сегментация.

### 3. Сверточные нейронные сети (CNN)

CNN являются основой многих успешных подходов в компьютерном зрении. Они специально разработаны для обработки изображений, используя свертки для выделения важных признаков.

## Обработка изображенй

```python
img = plt.imread("Datasets/SQL-meme.jpg")
print(img.shape)  # Это цветное изображение размером 579 на 1280, имеющее 3 канала для цвета - RGB
plt.imshow(img)
plt.show()
```

## Обрезка изображений

```python
img2 = img[100:800, 400:1000]
plt.imshow(img2)
plt.show()
```

## Увеличение яркости

```python
BrightnessUp = 100 * np.ones((img.shape), dtype='int32')
img3 = img + BrightnessUp
plt.imshow(img3)
plt.show()
```

## Снижение яркости

```python
BrightnessDown = -100 * np.ones((img.shape), dtype='int32')
img3 = img + BrightnessDown
plt.imshow(img3)
plt.show()
```

## Добавление шума на изображение

```python
AddNoise = np.random.randint(0, 100, img.shape)  # Сгенерируем случайное число в промежутке от 0 до 100 для параметра шума
img4 = img + AddNoise
plt.imshow(img4)
plt.show()
```

# Отобразим изображение с помощью OpenCV

```python
import cv2
```

## Прочитаем изображение, используя OpenCV

```python
imgOCV = cv2.imread("Datasets/SQL-meme.jpg")
plt.imshow(imgOCV)
plt.show()
```

## Преобразуем изображение, считанное OpenCV в формате BRG, в формат RGB

```python
imgOCV1 = cv2.cvtColor(imgOCV, cv2.COLOR_BGR2RGBA)
plt.imshow(imgOCV1)
plt.show()
```

## Считывание изображения в оттенках серого с помощью OpenCV

```python
imgGray = cv2.cvtColor(imgOCV, cv2.COLOR_BGR2GRAY)
plt.imshow(imgGray, cmap='gray')
plt.title('Черно-белое изображение')
plt.show()
```

## Определение границ объектов на изображении

```python
imgOCV3 = cv2.Canny(imgOCV, 150, 255)
plt.imshow(imgOCV3, cmap='gray')
plt.show()
```

## Размытие изображения

```python
# Отобразим базовое изображение
plt.imshow(cv2.cvtColor(imgOCV, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.show()

# Изменим размер изображения (уменьшим в 0,5 раза в обоих измерениях)
shrunken_img = cv2.resize(imgOCV, None, fx=0.1, fy=0.1, interpolation=cv2.INTER_AREA)

# Отобразим уменьшенное изображение
plt.imshow(cv2.cvtColor(shrunken_img, cv2.COLOR_BGR2RGB))
plt.title('Изображение со сниженным качеством')
plt.show()
```

```python
# Размытие сглаживает изображение
imgOCV8 = imgOCV1

blur = cv2.blur(imgOCV8, (18, 18))
gblur = cv2.GaussianBlur(imgOCV8, (7, 7), 0)

plt.imshow(imgOCV8), plt.title('Оригинальное изображение')
plt.show()
plt.imshow(blur), plt.title('Размытое изображение')
plt.show()
plt.imshow(gblur), plt.title('Изображение, размытое гауссианом')
plt.show()
```
