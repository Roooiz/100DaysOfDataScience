Прогноз временного ряда - это способ предсказания будущих значений временных данных на основе их предыдущих значений. Временной ряд представляет собой последовательность данных, измеренных в разные моменты времени. Это может быть что угодно: температура каждый день, продажи товаров каждый месяц, количество посетителей в интернет-магазине каждый час и так далее.

Прогноз временного ряда важен для многих областей, таких как финансы, экономика, метеорология, и многих других. Есть несколько методов для прогнозирования временных рядов, и вот некоторые из них:

1. **Метод сглаживания**: Этот метод основан на усреднении предыдущих значений временного ряда. Например, простое скользящее среднее или экспоненциальное сглаживание.

2. **Метод тренда и сезонности**: Временной ряд может включать в себя тренд (постепенное изменение во времени) и сезонность (циклические изменения). Прогнозирование может включать в себя учет этих факторов для более точных предсказаний.

3. **ARIMA-модели**: ARIMA (авторегрессия, интеграция, скользящее среднее) - это статистическая модель, которая учитывает автокорреляцию в данных и может быть эффективной для прогнозирования временных рядов.

4. **Машинное обучение**: Современные методы машинного обучения, такие как нейронные сети и алгоритмы глубокого обучения, также используются для прогнозирования временных рядов. Они могут автоматически извлекать сложные закономерности из данных.

Выбор метода зависит от характера данных и конкретной задачи прогнозирования. Важно также учитывать наличие выбросов, аномалий и других факторов, которые могут влиять на точность прогноза.

Расммотрим теоретическое наполнение каждого из ARIMA методов.
### AR (авторегрессия):
ARIMA состоит из трех основных компонентов: AR (авторегрессия), I (интеграция) и MA (скользящее среднее).

1. **AR (авторегрессия)**: Этот компонент отражает зависимость текущего значения временного ряда от предыдущих значений. Если мы говорим об AR(p), то это означает, что текущее значение зависит от предыдущих p значений.

    **Формула AR(p):**
    
    $ Y_t = \phi_1 \cdot Y_{t-1} + \phi_2 \cdot Y_{t-2} + \ldots + \phi_p \cdot Y_{t-p} + \epsilon_t $

    Где:
    - $Y_t$ - текущее значение временного ряда.
    - $\phi_1, \phi_2, \ldots, \phi_p$ - параметры авторегрессии.
    - $\epsilon_t$ - ошибка (случайная компонента) в момент времени $t$.

### MA (скользящее среднее):
2. **MA (скользящее среднее)**: Этот компонент отражает зависимость текущего значения временного ряда от предыдущих случайных ошибок.

    **Формула MA(q):**
    
    $ Y_t = \epsilon_t + \theta_1 \cdot \epsilon_{t-1} + \theta_2 \cdot \epsilon_{t-2} + \ldots + \theta_q \cdot \epsilon_{t-q} $

    Где:
    - $Y_t$ - текущее значение временного ряда.
    - $\epsilon_t$ - ошибка (случайная компонента) в момент времени $t$.
    - $\theta_1, \theta_2, \ldots, \theta_q$ - параметры скользящего среднего.

### ARMA (авторегрессия + скользящее среднее):
3. **ARMA (авторегрессия + скользящее среднее)**: Этот метод объединяет AR и MA компоненты.

    **Формула ARMA(p, q):**
    
    $ Y_t = \phi_1 \cdot Y_{t-1} + \phi_2 \cdot Y_{t-2} + \ldots + \phi_p \cdot Y_{t-p} + \epsilon_t + \theta_1 \cdot \epsilon_{t-1} + \theta_2 \cdot \epsilon_{t-2} + \ldots + \theta_q \cdot \epsilon_{t-q} $

### ARIMA (авторегрессия + интеграция + скользящее среднее):
4. **ARIMA (авторегрессия + интеграция + скользящее среднее)**: Этот метод добавляет компонент интеграции, который учитывает разницу между текущим значением и предыдущим значением временного ряда (дифференцирование).

    **Формула ARIMA(p, d, q):**
    
    $ Y_t = \phi_1 \cdot (Y_{t-1} - Y_{t-2}) + \phi_2 \cdot (Y_{t-2} - Y_{t-3}) + \ldots + \phi_p \cdot (Y_{t-p} - Y_{t-p-1}) + \epsilon_t + \theta_1 \cdot \epsilon_{t-1} + \theta_2 \cdot \epsilon_{t-2} + \ldots + \theta_q \cdot \epsilon_{t-q} $

    Где:
    - $d$ - порядок интеграции.

### SARIMA (сезонная ARIMA):
5. **SARIMA (сезонная ARIMA)**: Этот метод расширяет ARIMA для учета сезонности в данных. Он включает сезонные компоненты, такие как сезонная авторегрессия (SAR), сезонное интегрирование (I) и сезонное скользящее среднее (SMA).

    **Формула SARIMA(p, d, q)(P, D, Q)s:**
    
    $ Y_t = \phi_1 \cdot (Y_{t-1} - Y_{t-1-s}) + \ldots + \phi_p \cdot (Y_{t-p} - Y_{t-p-s}) + \epsilon_t + \theta_1 \cdot \epsilon_{t-1} + \ldots + \theta_q \cdot \epsilon_{t-q} $

    Где:
    - $P$ - порядок сезонной авторегрессии.
    - $D$ - порядок сезонного интегрирования.
    - $Q$ - порядок сезонного скользящего среднего.
    - $s$ - период сезонности.

### SARIMAX (сезонная ARIMA с внешними регрессорами):
6. **SARIMAX (сезонная ARIMA с внешними регрессорами)**: Этот метод расширяет SARIMA, добавляя внешние регрессоры, которые могут улучшить точность прогноза.

    **Формула SARIMAX(p, d, q)(P, D, Q)s:**
    
    $ Y_t = \beta \cdot X_t + \phi_1 \cdot (Y_{t-1} - Y_{t-1-s}) + \ldots + \phi_p \cdot (Y_{t-p} - Y_{t-p-s}) + \epsilon_t + \theta_1 \cdot \epsilon_{t-1} + \ldots + \theta_q \cdot \epsilon_{t-q} $

    Где:
    - $X_t$ - внешний регрессор.
    - $\beta$ - параметры внешнего регрессора.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv('Datasets/AirPassengers.csv')
data.head()
```

```python
data = data.rename(columns={"#Passengers": "Passengers"}, inplace=False)
data.head()
```

```python
data.columns = ['Month', 'Passengers']
data['Month'] = pd.to_datetime(data['Month'], format='%Y-%m')
data = data.set_index('Month')
data.head()
```

```python
data.plot(y='Passengers', figsize=(20, 4), color='blue', linestyle='-', marker='o', markersize=5,
          label='Пассажиропоток')
plt.grid(True)
plt.legend(loc='best')
plt.title('Кол-во пассажиров авиакомпании ')
plt.xlabel('Даты')
plt.ylabel('Кол-во пассажиров')
plt.show(block=False)
```

```python
train_len = 120
train = data[:train_len]
test = data[train_len:]
```

# Методы Авторегрессии

## Стационарный и нестационарный временной ряд

```python
data['Passengers'].plot(figsize=(12, 4))
plt.grid()
plt.legend(loc='best')
plt.title('Пассажиропоток авиакомпаний')
plt.show(block=False)
```

## Расширенный тест Дики-Фуллера (ADF)
H0: Временной ряд является нестационарным. Другими словами, он имеет некоторую структуру, зависящую от времени, и не имеет постоянной дисперсии во времени.
H1: временной ряд является стационарным.
Если p-значение из теста меньше некоторого уровня значимости (например, α = 0,05), то мы можем отвергнуть нулевую гипотезу и сделать вывод, что временной ряд является стационарным.

```python
from statsmodels.tsa.stattools import adfuller

adf_test = adfuller(data['Passengers'])

print('ADF статистика: %f' % adf_test[0])
print('Критическое значение @ 0.05: %.2f' % adf_test[4]['5%'])
print('p-value: %f' % adf_test[1])
```

## Тест Квятковского-Филлипса-Шмидта-Шина (KPSS)
В эконометрике тесты Квятковского-Филлипса-Шмидта-Шина используются для проверки нулевой гипотезы о том, что наблюдаемый временной ряд стационарен вокруг детерминированного тренда по сравнению с альтернативой в виде единичного корня.

```python
from statsmodels.tsa.stattools import kpss

kpss_test = kpss(data['Passengers'])

print('KPSS статистика: %f' % kpss_test[0])
print('CКритическое значение @ 0.05: %.2f' % kpss_test[3]['5%'])
print('p-value: %f' % kpss_test[1])
```

## Преобразование Бокса Кокса для обеспечения постоянной дисперсии

Широко используемый метод преобразования ненормально распределенного набора данных в более нормально распределенный.
Основная идея этого метода состоит в том, чтобы найти некоторое значение λ, чтобы преобразованные данные были как можно ближе к нормальному распределению, используя следующую формулу:
y(λ) = (y λ – 1) / λ, если y ≠ 0
y(λ) = log(y), если y = 0

```python
from scipy.stats import boxcox

data_boxcox = pd.Series(boxcox(data['Passengers'], lmbda=0), index=data.index)

plt.figure(figsize=(12, 4))
plt.grid()
plt.plot(data_boxcox, label='Данные после изменения')
plt.legend(loc='best')
plt.title('Временной ряд после использования метода Бокса Кокса')
plt.show()
```

## Дифференцируем ряд для устранения тренда

```python
data_boxcox_diff = pd.Series(data_boxcox - data_boxcox.shift(), data.index)
plt.figure(figsize=(12, 4))
plt.grid()
plt.plot(data_boxcox_diff, label='Измененные данные после дифференцирования')
plt.legend(loc='best')
plt.title('Измененный временной ряд после дифференцирования')
plt.show()
```

```python
data_boxcox_diff.dropna(inplace=True)
data_boxcox_diff.tail()
```

## Расширенный тест Дики-Фуллера (ADF) (проведем снова, после всех наших преобразований)

```python
adf_test = adfuller(data_boxcox_diff)

print('ADF статистика: %f' % adf_test[0])
print('Критическое значение @ 0.05: %.2f' % adf_test[4]['5%'])
print('p-value: %f' % adf_test[1])
```

## Тест Квятковского-Филлипса-Шмидта-Шина (KPSS)

```python
kpss_test = kpss(data_boxcox_diff)

print('KPSS статистика: %f' % kpss_test[0])
print('Критическое значение @ 0.05: %.2f' % kpss_test[3]['5%'])
print('p-value: %f' % kpss_test[1])
```

## Автокорреляционная функция (ACF)
Автокорреляционная функция - это характеристика сигнала, которая помогает находить повторяющиеся участки сигнала или определять несущую частоту сигнала, скрытую из-за наложений шума и колебаний на других частотах. Автокорреляционная функция часто используется в обработке сигналов и анализе временных рядов.
Неформально автокорреляционная функция - это сходство между значениями сигнала как функция от разницы во времени между ними.

```python
from statsmodels.graphics.tsaplots import plot_acf

plt.figure(figsize=(12, 4))
plt.grid()
plot_acf(data_boxcox_diff, ax=plt.gca(), lags=30)
plt.show()
```

## Частичная автокорреляционная функция (PACF)
При анализе временных рядов частичная автокорреляционная функция дает частичную корреляцию стационарного временного ряда с его собственными запаздывающими значениями, регрессируя значения временного ряда при всех более коротких задержках. Это контрастирует с автокорреляционной функцией, которая не контролирует другие задержки.

```python
from statsmodels.graphics.tsaplots import plot_pacf

plt.figure(figsize=(12, 4))
plt.grid()
plot_pacf(data_boxcox_diff, ax=plt.gca(), lags=30)
plt.show()
```

# Разделим набор данных на тренировочный и тестовый наборы

```python
train_data_boxcox = data_boxcox[:train_len]
test_data_boxcox = data_boxcox[train_len:]
train_data_boxcox_diff = data_boxcox_diff[:train_len - 1]
test_data_boxcox_diff = data_boxcox_diff[train_len - 1:]
```

```python
train_data_boxcox_diff[:10]
```

# Метод авторегрессии (AR)

```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(train_data_boxcox_diff, order=(1, 0, 0))
model_fit = model.fit()
model_fit.params
```

## Восстановим исходный временной ряд

```python
y_hat_ar = data_boxcox_diff.copy()
y_hat_ar['ar_forecast_boxcox_diff'] = model_fit.predict(data_boxcox_diff.index.min(), data_boxcox_diff.index.max())
y_hat_ar['ar_forecast_boxcox'] = y_hat_ar['ar_forecast_boxcox_diff'].cumsum()
y_hat_ar['ar_forecast_boxcox'] = y_hat_ar['ar_forecast_boxcox'].add(data_boxcox[0])
y_hat_ar['ar_forecast'] = np.exp(y_hat_ar['ar_forecast_boxcox'])
```

```python
plt.figure(figsize=(12, 4))
plt.grid()
plt.plot(train['Passengers'], label='Тренировочный набор')
plt.plot(test['Passengers'], label='Тестовый набор')
plt.plot(y_hat_ar['ar_forecast'][test.index.min():], label='Прогноз авторегрессии')
plt.legend(loc='best')
plt.title('Метод авторегрессии')
plt.show()
```

```python
rmse = np.sqrt(mean_squared_error(test['Passengers'], y_hat_ar['ar_forecast'][test.index.min():])).round(2)
mape = np.round(
    np.mean(np.abs(test['Passengers'] - y_hat_ar['ar_forecast'][test.index.min():]) / test['Passengers']) * 100, 2)
```

```python
results = pd.DataFrame(columns=['Method', 'RMSE', 'MAPE'])
tempResults = pd.DataFrame({'Method': ['Autoregressive (AR) method'], 'RMSE': [rmse], 'MAPE': [mape]})
results = pd.concat([results, tempResults])
results = results[['Method', 'RMSE', 'MAPE']]
results
```
